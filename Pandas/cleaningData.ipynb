```python
#cleaning data
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")

#handling missing values

df.isnull() #df.isna()

df.isnull().sum()

df.dropna() #row drop
df.dropna(axis = 1) #col drop
df.fillna(0)

age_mean = df["age"].mean()
cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)
cleaned_data

df.ffill()
df.bfill()
```

```python
#handling duplicate values

df.duplicated()
df2 = df.copy()
df.drop_duplicates()
```
