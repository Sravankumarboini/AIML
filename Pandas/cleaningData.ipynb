```python
#cleaning data
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")

#handling missing values

df.isnull() #df.isna()

df.isnull().sum()

df.dropna() #row drop
df.dropna(axis = 1) #col drop
df.fillna(0)

age_mean = df["age"].mean()
cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)
cleaned_data

df.ffill()
df.bfill()
```

```python
#handling duplicate values

df.duplicated()
df2 = df.copy()
df.drop_duplicates()
```

```python
#handle data types & date-time

df.dtypes
df2 = df.copy()
df2 = df2.fillna(0)
df2 = df2["age"].astype("int64")
df2.dtypes
```

```python
#handle stirngs
df["gender"].str.lower()
df["gender"].str.lower()
df["gender"].str.capitalize()

df["name"].str.split(" ")

df["country"].str.contains("india", case = False)
```

```python
#Tronsforming Data(Feature Engineering)
df2 = df.copy()

df2["tax"] = df["income"].apply(lambda x : "20%" if x >= 50000 else "10%")

gender_map = {"Male" : "M", "Female":"F", "Unknown" : "U"}

df2["gender"] = df2["gender"].map(gender_map)

df2 = df2.assign(new_income = df2["income"] * 1.1)

df2["country"] = df2["country"].replace("USA", "US")
df2
```
