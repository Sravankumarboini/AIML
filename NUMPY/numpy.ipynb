```python
import numpy as np
import time
```

```python
#execution performance
size = 1_000_000

#python list
py_list = list(range(size))
start = time.time()
sq_list = [i**2 for i in py_list]
end = time.time()
print(f"python list time = {end-start} seconds")

#numpy array
np_array = np.array(py_list)
start = time.time()
sq_array = np_array**2
end = time.time()
print(f"nummpy array time = {end-start} seconds")
```

```python
#memory 
import sys

print(f"python list size = {sys.getsizeof(py_list)} bytes")

print(f"numpy array size = {np_array.nbytes} bytes")
```
