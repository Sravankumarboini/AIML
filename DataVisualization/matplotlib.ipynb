```python
#Real data

oscar_movie = [
    "The dark knight",
    "The hurt locker",
    "The kings speech",
    "The artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] #$M

#plot
plt.plot(years, oscar_revenue)
plt.title("Oscar movies revenue in each year")
plt.xlabel("years")
plt.ylabel("Revenue in $M")
```

```python
non_oscar_movies = [
    "slumdogmillionaire",
    "avatar",
    "incaption",
    "higo",
    "lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275] #$M

plt.plot(years, oscar_revenue, label="oscar movies")
plt.plot(years, non_oscar_revenue, label="non oscar movies")

plt.title("oscar movies vs non oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenus in $M")

plt.legend()
```

```python
plt.plot(years, oscar_revenue,"o--r", label="oscar movies")
plt.plot(years, non_oscar_revenue,"<-y", label="non oscar movies")

plt.title("oscar movies vs non oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenus in $M")

plt.legend()
```

```python
plt.plot(years, oscar_revenue, label="oscar movies")
plt.plot(years, non_oscar_revenue, label="non oscar movies")

plt.title("oscar movies vs non oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenus in $M")

plt.legend()
plt.style.available
plt.style.use("dark_background")
```
