```python
#Real data

oscar_movie = [
    "The dark knight",
    "The hurt locker",
    "The kings speech",
    "The artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] #$M

#plot
plt.plot(years, oscar_revenue)
plt.title("Oscar movies revenue in each year")
plt.xlabel("years")
plt.ylabel("Revenue in $M")
```

```python
non_oscar_movies = [
    "slumdogmillionaire",
    "avatar",
    "incaption",
    "higo",
    "lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275] #$M

plt.plot(years, oscar_revenue, label="oscar movies")
plt.plot(years, non_oscar_revenue, label="non oscar movies")

plt.title("oscar movies vs non oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenus in $M")

plt.legend()
```

```python
plt.plot(years, oscar_revenue,"o--r", label="oscar movies")
plt.plot(years, non_oscar_revenue,"<-y", label="non oscar movies")

plt.title("oscar movies vs non oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenus in $M")

plt.legend()
```

```python
plt.plot(years, oscar_revenue, label="oscar movies")
plt.plot(years, non_oscar_revenue, label="non oscar movies")

plt.title("oscar movies vs non oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenus in $M")

plt.legend()
plt.style.available
plt.style.use("dark_background")
```

```python
#movies - revenue

plt.style.use("default")
plt.plot(oscar_movie, oscar_revenue)
plt.title("movies revenue")
plt.xlabel("movies")
plt.ylabel("revenue in $M")
plt.tight_layout()
```

```python
#Bar plot

plt.bar(years, oscar_revenue, color="green")

plt.title("revenue in each year for oscar movies")
plt.xlabel("years")
plt.ylabel("revenue in $M")

for i in range(len(years)): 
    plt.text(years[i], oscar_revenue[i]+10, oscar_revenue[0], ha="center")

plt.ylim(0, max(oscar_revenue)+100)
```

```python
#oscar vs non oscar movies
import numpy as np

width= 0.4
x = np.arange(len(years))

plt.title("oscar va non oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenue in $M")
plt.bar(x-width/2, oscar_revenue, width, label="oscar movies")
plt.bar(x+width/2, non_oscar_revenue, width, label="non oscar movies")

plt.legend()
plt.xticks(x, years)
```
