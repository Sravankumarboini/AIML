```python
#Real data

oscar_movie = [
    "The dark knight",
    "The hurt locker",
    "The kings speech",
    "The artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] #$M

#plot
plt.plot(years, oscar_revenue)
plt.title("Oscar movies revenue in each year")
plt.xlabel("years")
plt.ylabel("Revenue in $M")
```

```python
non_oscar_movies = [
    "slumdogmillionaire",
    "avatar",
    "incaption",
    "higo",
    "lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275] #$M

plt.plot(years, oscar_revenue, label="oscar movies")
plt.plot(years, non_oscar_revenue, label="non oscar movies")

plt.title("oscar movies vs non oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenus in $M")

plt.legend()
```

```python
plt.plot(years, oscar_revenue,"o--r", label="oscar movies")
plt.plot(years, non_oscar_revenue,"<-y", label="non oscar movies")

plt.title("oscar movies vs non oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenus in $M")

plt.legend()
```

```python
plt.plot(years, oscar_revenue, label="oscar movies")
plt.plot(years, non_oscar_revenue, label="non oscar movies")

plt.title("oscar movies vs non oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenus in $M")

plt.legend()
plt.style.available
plt.style.use("dark_background")
```

```python
#movies - revenue

plt.style.use("default")
plt.plot(oscar_movie, oscar_revenue)
plt.title("movies revenue")
plt.xlabel("movies")
plt.ylabel("revenue in $M")
plt.tight_layout()
```

```python
#Bar plot

plt.bar(years, oscar_revenue, color="green")

plt.title("revenue in each year for oscar movies")
plt.xlabel("years")
plt.ylabel("revenue in $M")

for i in range(len(years)): 
    plt.text(years[i], oscar_revenue[i]+10, oscar_revenue[0], ha="center")

plt.ylim(0, max(oscar_revenue)+100)
```

```python
#oscar vs non oscar movies
import numpy as np

width= 0.4
x = np.arange(len(years))

plt.title("oscar va non oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenue in $M")
plt.bar(x-width/2, oscar_revenue, width, label="oscar movies")
plt.bar(x+width/2, non_oscar_revenue, width, label="non oscar movies")

plt.legend()
plt.xticks(x, years)
```

```python
#horizontal bar

plt.barh(oscar_movie, oscar_revenue)
```

```python
#scatter plot

people = ["person A", "person B", "person C", "person D", "person E",
         "person F", "person G", "person H", "person I", "person J",]

age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors = ["green" if x < 135 else "red" for x in blood_pressure]
plt.scatter(age, blood_pressure, cmap="OrRd",c=blood_pressure, alpha= 1)
plt.title("Age vs BP")
plt.xlabel("age")
plt.ylabel("Blood Pressure")

plt.grid()
plt.colorbar()

for i in range(len(people)):
    plt.annotate(people[i], xy=(age[i], blood_pressure[i]), xytext=(age[i]+1, blood_pressure[i]+1))

plt.xlim(min(age), max(age)+10)
plt.ylim(min(blood_pressure), max(blood_pressure)+5)
```

```python
#Histogram 

import numpy as np
np.random.seed(0)

scores = np.random.normal(70,10,100) #normal distribution

plt.hist(scores, color="skyblue", edgecolor="black", bins=[0,30,70,90,100])
plt.ylabel("# of students")
plt.xlabel("scores")
```

```python
#Multiple datasets

legit_transactions = [ 2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99,
                       39.50, 45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 99.99,
                       120.50, 150.00, 199.99, 249.99, 300.75, 450.00, 600.00,
                       850.00, 1200.00 ] 
fraud_transactions = [ 50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000, 1200, 1500,
                       1500, 2000, 2500, 3000, 3000]


#Transaction amount distribution -- legit vs fraud

plt.hist(fraud_transactions, color="red",bins=10, label="fraud", edgecolor="black", alpha=0.5)

plt.hist(legit_transactions, color="green",bins=10, label="legit", edgecolor="black", alpha=0.5)

plt.title("Transaction amount distribution -- legit vs fraud")
plt.xlabel("amount in $")
plt.ylabel("frequency")
plt.grid()

plt.legend()
```
